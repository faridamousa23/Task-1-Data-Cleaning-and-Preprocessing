```python
df = df.drop(columns =['All Time Peak','Peak'])
```

```python

df['Ref.'] = df['Ref.'].astype(str)
df['Ref.'] = df['Ref.'].str.replace(r'[\[\]]', '', regex=True)
df['Ref.'] = pd.to_numeric(df['Ref.'], errors='coerce').astype('Int64')
df
```

```python
import pandas as pd
import re
money_cols = ["Actual\xa0gross", "Adjusted\xa0gross (in 2022 dollars)", "Average gross"]
for col in money_cols:
    df[col] = (
        df[col].astype(str)
        .str.replace(r'[^0-9]', '', regex=True)
        .replace('', '0')
    )
    df[col] = pd.to_numeric(df[col])


df
```

```python
df["First Year"]=df['Year(s)'].str[:4]
```

```python
 df["last Year"]=df['Year(s)'].str[:-4].str.replace(r'[^0-9]','',regex=True)
```

```python
df.head()
```

```python
df = df.drop(columns =['Year(s)'])
```

```python
 df["last Year"]= df["last Year"].replace("", "0")
```

```python
df = df.dropna(subset=['Ref.'])
```

```python
df["Tour title"] = df["Tour title"].astype(str).str.replace(r'[\*\†‡\[\]a-z0-9]+$', '', regex=True)
df["Tour title"] = df["Tour title"].str.replace(r'\[[^\]]*\]', '', regex=True)
df["Tour title"] = df["Tour title"].str.strip()
```

```python
df["Rank"] = pd.to_numeric(df["Rank"], errors="coerce").astype("Int64")
df["Shows"] = pd.to_numeric(df["Shows"], errors="coerce").astype("Int64")
```
